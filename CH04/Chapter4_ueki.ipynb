# 4.1　移動平均(MA)プロセスを定義する

移動平均プロセス

現在の値が現在と過去の誤差項に線形従属する。

誤差項は互いに独立しており、正規分布である。＝ホワイトノイズ

移動平均モデルはMA(q)で表され、qは次数を表す。

次数qの一般的な移動平均プロセス

y(t)：現在の観測値、μ：時系列の平均、ε(t)：現在の誤差項、

ε(t-q)：過去の誤差項、係数θ(q)：過去の誤差が現在の値に与える影響の大きさ


$$
Y_t = \mu + \varepsilon_t + \theta_1 \varepsilon_{t-1} + \cdots + \theta_q \varepsilon_{t-q}
$$


次数が１＝MA(1)の場合

$$
Y_t = \mu + \varepsilon_t + \theta_1 \varepsilon_{t-1}
$$

$$
\mathbb{E}[Y_t] = \mu
$$

$$
\operatorname{Var}(Y_t)
= \sigma^2 \left(1 + \theta_1^2  \right)
$$

$$
\mathrm{Cov}(Y_t, Y_{t+1})
= \theta \sigma^2
$$

$$
\rho(1) = \frac{\mathrm{Cov}(Y_t, Y_{t+1})}{\operatorname{Var}(Y_t)} = \frac{\theta}{1 + \theta^2}, \quad \rho(h) = 0 \quad (h > 1)
$$



次数が２＝MA(2)の場合

$$
Y_t = \mu + \varepsilon_t + \theta_1 \varepsilon_{t-1} + \theta_2 \varepsilon_{t-2}
$$

$$
\mathbb{E}[Y_t] = \mu
$$

$$
\operatorname{Var}(Y_t) = \sigma^2 \left(1 + \theta_1^2 + \theta_2^2 \right)
$$


$$
\mathrm{Cov}(Y_t, Y_{t+2}) = \sigma^2 (\theta_1 + \theta_1 \theta_2)
$$


$$
\mathrm{Cov}(Y_t, Y_{t+2}) = \theta_2 \sigma^2
$$

$$
\rho(1) = \frac{\mathrm{Cov}(Y_t, Y_{t+1})}{\operatorname{Var}(Y_t)} = \frac{\theta_1 + \theta_1 \theta_2}{1 + \theta_1^2 + \theta_2^2}
$$

$$
\rho(2) = \frac{\mathrm{Cov}(Y_t, Y_{t+2})}{\operatorname{Var}(Y_t)} = \frac{\theta_2}{1 + \theta_1^2 + \theta_2^2}
$$

$$
\rho(h) = 0 \quad (h > 2)
$$


MA(q)

$$
Y_t = \mu + \varepsilon_t + \theta_1 \varepsilon_{t-1} + \theta_2 \varepsilon_{t-2} + \cdots + \theta_q \varepsilon_{t-q}
$$

$$
\mathbb{E}[Y_t] = \mu
$$

$$
\operatorname{Var}(Y_t) = \sigma^2 \left(1 + \theta_1^2 + \theta_2^2 + \cdots + \theta_q^2 \right)
$$

$$
\mathrm{Cov}(Y_t, Y_{t+h}) = \sigma^2 \sum_{i=0}^{q-h} \theta_i \theta_{i+h} \quad \text{for } h \leq q
$$

$$
\mathrm{Cov}(Y_t, Y_{t+h}) = 0 \quad \text{for } h > q
$$

$$
\rho(h) = \frac{\mathrm{Cov}(Y_t, Y_{t+h})}{\operatorname{Var}(Y_t)} =
\frac{\sum_{i=0}^{q-h} \theta_i \theta_{i+h}}{1 + \theta_1^2 + \theta_2^2 + \cdots + \theta_q^2} \quad \text{for } h \leq q
$$

$$
\rho(h) = 0 \quad \text{for } h > q
$$



MA(q) モデルの分散の導出

$$
\mathrm{Var}(Y_t) = \mathbb{E}[Y_t^2] - (\mathbb{E}[Y_t])^2
$$

$
(\mathbb{E}[Y_t] = \mu)なので
$

$$
\mathrm{Var}(Y_t) = \mathbb{E}[Y_t^2] - \mu^2
$$

$
Y_t^2 = \left( \mu + \varepsilon_t + \theta_1 \varepsilon_{t-1} + \theta_2 \varepsilon_{t-2} + \cdots + \theta_q \varepsilon_{t-q} \right)^2
$
$$
= \mu^2 + \varepsilon_t^2 + \theta_1^2 \varepsilon_{t-1}^2 + \cdots + \theta_q^2 \varepsilon_{t-q}^2 + 2 \mu \varepsilon_t + 2 \mu \theta_1 \varepsilon_{t-1} + \cdots + 2 \mu \theta_q \varepsilon_{t-q} + \text{クロス項}
$$

ホワイトノイズの特性を利用して、クロス項の期待値はゼロになります。残る項は次の通りです：

$$
\mathbb{E}[Y_t^2] = \mu^2 + \sigma^2 \left(1 + \theta_1^2 + \theta_2^2 + \cdots + \theta_q^2 \right)
$$

したがって、分散は次のように求められます：

$$
\mathrm{Var}(Y_t) = \sigma^2 \left(1 + \theta_1^2 + \theta_2^2 + \cdots + \theta_q^2 \right)
$$


MA(q) モデルの共分散、相関係数の導出

$
h \leq q \text{ のとき、}
$
$$
\mathrm{Cov}(Y_t, Y_{t+h}) = \mathbb{E}\left[\left(\sum_{i=0}^q \theta_i \varepsilon_{t-i} \right) \left( \sum_{j=0}^q \theta_j \varepsilon_{t+h-j} \right)\right]
$$

$$
= \sum_{i=0}^q \sum_{j=0}^q \theta_i \theta_j \mathbb{E}[\varepsilon_{t-i} \varepsilon_{t+h-j}]
$$

$$
t - i = t + h - j \quad \Rightarrow \quad j = i + h
$$

$$
\sum_{i=0}^q \sum_{j=0}^q \theta_i \theta_j \mathbb{E}[\varepsilon_{t-i} \varepsilon_{t+h-j}]
\quad \Rightarrow \quad
\sum_{i=0}^q \theta_i \theta_{i+h} \cdot \mathbb{E}[\varepsilon_{t-i} \varepsilon_{t-i}]
$$

$$
= \sigma^2 \sum_{i=0}^{q-h} \theta_i \theta_{i+h} \quad
$$

$$
\rho(h) = \frac{\mathrm{Cov}(Y_t, Y_{t+h})}{\operatorname{Var}(Y_t)}
$$

$$
= \frac{\sum_{i=0}^{q-h} \theta_i \theta_{i+h}}{1 + \theta_1^2 + \theta_2^2 + \cdots + \theta_q^2}
$$


$$
\mathrm{Cov}(Y_t, Y_{t+h}) = 0 \quad\ \rho(h) = 0 \quad\text{ for } h > q \text{}
$$



qの値が大きければ大きいほど、現在の値に影響を与える過去の誤差項の個数は増えていく。

MA(2)モデルによる予測

$$
Y_t = \mu + \varepsilon_t + \theta_1 \varepsilon_{t-1} + \theta_2 \varepsilon_{t-2}
$$

パラメーター
$
　\mu , 　\theta_1  , 　 \theta_2
$

は最大尤度推定（MLE）によって決定する。

$$
\varepsilon_t = Y_t - \mu - \theta_1 \varepsilon_{t-1} - \theta_2 \varepsilon_{t-2}
$$

$$
f(\varepsilon_t) = \frac{1}{\sqrt{2\pi \sigma^2}} \exp\left( -\frac{\varepsilon_t^2}{2\sigma^2} \right)
$$

$$
\mathcal{L} = \prod_{t=3}^{T} \frac{1}{\sqrt{2\pi \sigma^2}} \exp\left( -\frac{1}{2\sigma^2} \left( Y_t - \mu - \theta_1 \varepsilon_{t-1} - \theta_2 \varepsilon_{t-2} \right)^2 \right)
$$

最初の 2 時点（t=1 と t=2）については、誤差を計算できないためt=3 から 始める。

この尤度が最大にパラメーターを探す。

$$
\log \mathcal{L} = -\frac{T-2}{2} \log(2\pi \sigma^2) - \frac{1}{2\sigma^2} \sum_{t=3}^{T} \left( Y_t - \mu - \theta_1 \varepsilon_{t-1} - \theta_2 \varepsilon_{t-2} \right)^2
$$

$
\muに関する偏微分：
$
$$
\frac{\partial}{\partial \mu} \log \mathcal{L} = \frac{1}{\sigma^2} \sum_{t=3}^{T} \left( Y_t - \mu - \theta_1 \varepsilon_{t-1} - \theta_2 \varepsilon_{t-2} \right)
$$

$
\theta_1 に関する偏微分：
$

$$
\frac{\partial}{\partial \theta_1} \log \mathcal{L} = \frac{1}{\sigma^2} \sum_{t=3}^{T} \left( Y_t - \mu - \theta_1 \varepsilon_{t-1} - \theta_2 \varepsilon_{t-2} \right) \varepsilon_{t-1}
$$

$
\theta_2\ に関する偏微分：
$

$$
\frac{\partial}{\partial \theta_2} \log \mathcal{L} = \frac{1}{\sigma^2} \sum_{t=3}^{T} \left( Y_t - \mu - \theta_1 \varepsilon_{t-1} - \theta_2 \varepsilon_{t-2} \right) \varepsilon_{t-2}
$$

$
\sigma^2に関する偏微分：
$

$$
\frac{\partial}{\partial \sigma^2} \log \mathcal{L} = -\frac{T-2}{2\sigma^2} + \frac{1}{2 (\sigma^2)^2} \sum_{t=3}^{T} \varepsilon_t^2
$$

これらの偏微分が0になる点が最適なパラメーター


MA(2)では、予測値は次のように表される。

$$
Y_t = \mu + \theta_1 \varepsilon_{t-1} + \theta_2 \varepsilon_{t-2}
$$

MA(q)モデルで予測できるのはqステップ先まで
それより先は過去の誤差項がないため、平均値が返される。

回避するためにはローリング予測を行う。

ローリング予測

「現時点までのデータ」で予測できるだけの未来を予測し、
その予測結果を使って、さらにその先を予測していく方法。


ローリング予測を行う関数

````python
from statsmodels.tsa.statespace.sarimax import SARIMAX

def rolling_forecast(df: pd.DataFrame,train_len: int, horizon: int,
                     window: int, method: str) -> list:

    total_len = train_len + horizon

    if method == 'mean':
        pred_mean = []

        for i in range(train_len, total_len, window):
            mean = np.mean(df[:i].values)
            pred_mean.extend(mean for _ in range(window))

        return pred_mean

    elif method == 'last':
        pred_last_value = []

        for i in range(train_len, total_len, window):
            last_value = df[:i].iloc[-1].values[0]
            pred_last_value.extend(last_value for _ in range(window))

        return pred_last_value

    elif method == 'MA':
        pred_MA = []

        for i in range(train_len, total_len, window):
            model = SARIMAX(df[:i], order=(0,0,2))
            res = model.fit(disp=False)
            predictions = res.get_prediction(0, i + window - 1)
            oos_pred = predictions.predicted_mean.iloc[-window:]
            pred_MA.extend(oos_pred)

        return pred_MA
````

この関数を用いて予測を行う

````python
pred_df = test.copy()

TRAIN_LEN = len(train)
HORIZON = len(test)
WINDOW = 2

pred_mean = rolling_forecast(df_diff, TRAIN_LEN, HORIZON, WINDOW, 'mean')
pred_last_value = rolling_forecast(df_diff, TRAIN_LEN, HORIZON, WINDOW, 'last')
pred_MA = rolling_forecast(df_diff, TRAIN_LEN, HORIZON, WINDOW, 'MA')

pred_df['pred_mean'] = pred_mean
pred_df['pred_last_value'] = pred_last_value
pred_df['pred_MA'] = pred_MA

pred_df.head()
````

# 4.4　実習

# 4.4.1　MA(2)プロセスのシミュレーションと予測

````python
from statsmodels.tsa.arima_process import ArmaProcess
import numpy as np

np.random.seed(42)

ma2 = np.array([1, 0.9, 0.3])   # MA(2): θ1 = 0.9, θ2 = 0.3
ar2 = np.array([1, 0, 0])   # AR(0) なので自己回帰なし（定数項1のみ）

MA2_process = ArmaProcess(ar2, ma2).generate_sample(nsample=1000)
````

````python
fig, ax = plt.subplots()

ax.plot(MA2_process)
ax.set_xlabel('Timesteps')
ax.set_ylabel('Value')

fig.autofmt_xdate()
plt.tight_layout()
````

````python
ADF_result = adfuller(MA2_process)

print(f'ADF Statistic: {ADF_result[0]}')
print(f'p-value: {ADF_result[1]}')
````

````python
plot_acf(MA2_process, lags=30)
plt.tight_layout()
````

````python
df = pd.DataFrame({'value':MA2_process})

train = df[:int(0.8*len(df))]
test = df[int(0.8*len(df)):]

print(len(train))
print(len(test))
````

````python
TRAIN_LEN = len(train)
HORIZON = len(test)
WINDOW = 2

pred_mean = rolling_forecast(df, TRAIN_LEN, HORIZON, WINDOW, 'mean')
pred_last_value = rolling_forecast(df, TRAIN_LEN, HORIZON, WINDOW, 'last')
pred_MA2 = rolling_forecast(df, TRAIN_LEN, HORIZON, WINDOW, 'MA')

test.loc[:, 'pred_mean'] = pred_mean
test.loc[:, 'pred_last_value'] = pred_last_value
test.loc[:, 'pred_MA2'] = pred_MA2

test.head()
````
