## AR過程
　　⇒時点tにおけるモデル出力が時刻t以前のモデル出力に依存する確率過程

$$
y_{t}=C+\phi_{1}\ y_{t-1}+\phi_{2}\ y_{t-2}+・・・+\phi_{p}\ y_{t-p}+\epsilon_{t}
$$
$\epsilon_{t}$は平均0分散$\sigma^{2}$のガウス分布に従い、$y_{t}$の平均は$\mu$とする。

AR(1)の平均・分散・自己共分散
$$
y_{t}=C+\phi_{1}\ y_{t-1}+\epsilon_{t}
$$
$$
\mu=C+\phi_{1}\mu
$$
$$
\mu=\frac{C}{1-\phi_{1}}
$$

分散を$\gamma_{0}$とすると
$$
\gamma_{0}=\phi_{1}^{2}\gamma_{0}+\sigma^{2}
$$
$$
\gamma_{0}=\frac{\sigma^{2}}{1-\phi_{1}^{2}}
$$
ここから$|\phi_{1}|<1$の時に分散が存在し、定常となることがわかる。

自己共分散を考える。<br>
ラグが0の時
$$\gamma(0)=var(y_{t})=\frac{\sigma^{2}}{1-\phi_{1}^{2}}$$<br>
ラグが1の時
$$\gamma(1)=Cov(y_{t},y_{t-1})=\phi_{1}Var(y_{t-1}+Cov(\epsilon_{t},y_{t-1}))$$
$$\gamma(1)=\phi_{1}\gamma(0)$$
ラグ2の時
$$\gamma(1)=Cov(y_{t},y_{t-2})=\phi_{1}Cov(y_{t-1},y_{t-2})+Cov(\epsilon_{t},y_{t-2}))$$
$$\gamma(2)=\phi_{1}^{2}\gamma(0)$$
一般的な時はYule-Walker方程式を使う。<br>
この方程式は以下のように示される。
$$\gamma(h)=\sum_{k=1}^{p}\phi_{k}\gamma(h-k)$$
AR(1)のときは
$$\gamma(h)=\phi_{1}\gamma(h-1)$$となるのでこの式を解くと
$$\gamma(h)=\phi_{1}^{h}\gamma(0)$$
ここで自己相関関数(ACF)を考える。これは自己共分散関数をラグの共分散で割ったもので表される。
$$\rho(h)=\frac{\gamma(h)}{\gamma(0)}$$
この時$\rho(0)=1$であり、$|\rho(h)|\leq1$。
時系列が定常ならば$\rho(h)$はhにのみ依存する。

## ラグ演算子
 ラグ演算子・・・時系列を1期過去にずらす演算
 $$Ly_{t}=y_{t-1},L^{2}y_{t}=y_{t-2},L^{k}y_{t}=y_{t-k}$$
 AR(1)の時
 $$y_{t}=C+\phi_{1}y_{t-1}+\epsilon_{t}=C+\phi_{1}Ly_{t}+\epsilon_{t}$$
 $$y_{t}=(1-\phi_{1}L)^{-1}(C+\epsilon_{t})$$
 等比数列の和の公式を用いると
 $$(1-\phi_{1}L)^{-1}=1+(\phi_{1}L)+(\phi_{1}L)^{2}+・・・$$
 また、Cに関しては時系列における定数であるので$LC=C$<br>
 以上のことより
 $$y_{t}=\frac{C}{1-\phi_{1}L}+\epsilon_{t}+\phi_{1}\epsilon_{t-1}+\phi_{1}^{2}\epsilon_{t-2}・・・$$
 よって定常となり、MA(q)過程のq→∞の場合と一致する（変換可能である）

## AR(2)の平均・分散・自己共分散
$$
y_{t}=C+\phi_{1}\ y_{t-1}+\phi_{2}\ y_{t-2}+\epsilon_{t}
$$
$$
\mu=C+\phi_{1}\mu+\phi_{2}\mu
$$
$$
\mu=\frac{C}{1-\phi_{1}-\phi_{2}}
$$
分散は
$$
\gamma_{0}=\phi_{1}^{2}\gamma(0)+\phi_{2}^{2}\gamma(0)+\sigma^{2}
$$
$$
\gamma_{0}=\frac{\sigma^{2}}{1-\phi_{1}^{2}-\phi_{2}^{2}}
$$

## 偏自己相関（PACF） 
  $$ \alpha(h) = \text{Corr}(X_t, X_{t-h} \mid X_{t-1}, \dots, X_{t-h+1})$$
  中間のラグの影響を取り除いた上での純粋な$X_t $と $X_{t-h}$の相関

## イメージで言うと…

- ACFで「ラグ3」が大きい相関を持っていたとしても、それは実はラグ1・2の影響を通じたものかもしれない。
- PACFは、ラグ1と2の影響を除いたあとで「本当にラグ3が影響しているのか？」を教えてくれる。

# 2. PACFはどんなときに使うの？

主に： ARモデルの次数 p を決めるとき

- ACFは**MAモデルの次数**に有用（急に切れる）
- PACFは**ARモデルの次数**に有用（急に切れる）

### 使い方の例：

- PACFをプロットしてみて、ラグ h で初めて相関が0になる → AR(p)モデルの候補は $ p = h - 1$
- 例：PACFがラグ3で急にゼロに落ちる → AR(2)が良さそう

# 3. PACFの数式定義と求め方

## 数式定義（厳密）

偏自己相関 $alpha(h)$ は、**線形回帰**で求まります。

以下の回帰式を考えます：

$$
X_t = \phi_{h1} X_{t-1} + \phi_{h2} X_{t-2} + \cdots + \phi_{hh} X_{t-h} + \epsilon_t^{(h)}
$$

このとき、$ X_t$ に対する $ X_{t-h} $の回帰係数（最後の係数）

$$
\boxed{\alpha(h) = \phi_{hh}}
$$

これが「ラグhにおける偏自己相関係数」です。

この係数は、**Yule-Walker方程式**や**Durbin-Levinsonアルゴリズム**を使って計算できます。

# 4. なぜAR(p)ではラグ \( h > p \) でPACFが0になるのか？

## AR(p) モデルとは？

$$
X_t = \phi_1 X_{t-1} + \cdots + \phi_p X_{t-p} + \epsilon_t
$$

- **ラグp までの過去しか使っていない**
- $ X_t $は $X_{t-1}, X_{t-2}, ..., X_{t-p}$の**線形結合** + ノイズ

## 結論：

- AR(p) 過程では、PACFは：
  $$
  \alpha(h) = 0 \quad \text{for all } h > p
  $$

つまり：
- ラグpまで → PACF ≠ 0（可能性あり）
- ラグpより先 → PACF = 0（理論的に）

これは理論的にも、実務的にも非常に強力な**モデル特定（model identification）**のヒントになります。

# 5.使い方まとめ

| ACF | PACF | モデル判断にどう使う？ |
|-----|------|--------------------------|
| ゆっくり減衰 | 急にゼロ（例：ラグ3以降ゼロ） | AR(2) モデルを疑う |
| 急にゼロ | 減衰 | MAモデルを疑う |
| 両方減衰 | ARMAモデルを疑う |

具体的に求める時には
Durbin-Levinsonアルゴリズムを使用する
$$
\phi_{k}^{(k)} = \frac{\gamma(k) - \sum_{j=1}^{k-1} \phi_{j}^{(k-1)} \gamma(k - j)}{\sigma_{k-1}^2}
$$

$$
\phi_{j}^{(k)} = \phi_{j}^{(k-1)} - \phi_{k}^{(k)} \phi_{k - j}^{(k-1)}, \quad j = 1, 2, ..., k - 1
$$

$$
\sigma_k^2 = \sigma_{k-1}^2 (1 - (\phi_k^{(k)})^2)
$$
